```
from sklearn.svm import LinearSVC
modelo = LinearSVC()
modelo.fit(train_x, train_y)
```

```
misterioso1 = [1,1,1]
misterioso2 = [1,1,0]
misterioso3 = [0,1,1]

test_x = [misterioso1,misterioso2,misterioso3]
test_y = [1,0,1]

pred = modelo.predict(test_x)

correctos = (pred == test_y).sum()
tot = len(test_y)
prop_aciertos = correctos/tot

print("% de acierto: ", prop_aciertos * 100)
```

```
from sklearn.metrics import accuracy_score
print("% de acierto: ", accuracy_score(test_y, pred) * 100)
print("Porcentaje de acierto %.2f " % (accuracy_score(test_y, pred) * 100))

```
